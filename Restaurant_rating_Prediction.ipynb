```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Convert only numeric columns
df_numeric = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = df_numeric.corr()

# Create heatmap
plt.figure(figsize=(6, 4))
import seaborn as sns
sns.heatmap(corr_matrix, annot=True, cmap="cividis")
plt.show()
```

```python
#applying standscaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#scaling
X = scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
```

```python
#evaluting model
def Model_results(Predictions):
    print("Mean absolute error on model is{}".format(mean_absolute_error(y_test,Predictions)))
    print("Root mean square error on model is {}".format(np.sqrt(mean_squared_error(y_test,Predictions))))
```

```python
#applying linearregression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#fitting model
lr.fit(X_train,y_train)
```

```python
Predictions_by_lr = lr.predict(X_test)
```

```python
Model_results(Predictions_by_lr)
```

```python
Predictions_by_lr
```

```python
#applying SVM
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
SVM_model = SVR()
```

```python
#defining a parameter grid for hyperparameter
param_gridsv = {"C" : [0.1,0.5],"kernel":["Linear","rbf"],"degree":[4]}
```

```python
gridsvr = GridSearchCV(SVM_model,param_gridsv)

#fitting the model 
gridsvr.fit(X_train,y_train)
```

```python
pred_gridsvr = gridsvr.predict(X_test)
```

```python
Model_results(pred_gridsvr)
```

```python
#applying decision_tree 
from sklearn.tree import DecisionTreeRegressor
```

```python
#hyperparameter tuning
param_grid={
    "max_depth" : [20,50],
    "min_samples_leaf" : [2,4],
    "min_samples_split" : [4,9]
}
```

```python
#fitting model
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator=tree_model,param_grid=param_grid)
```

```python
grid_tree.fit(X_train,y_train)
```

```python
tree_predictions =grid_tree.predict(X_test)
```

```python
Model_results(tree_predictions)
```

```python
#applying random_forest
from sklearn.ensemble import RandomForestRegressor
```

```python
#fitting model
r_model = RandomForestRegressor()
r_model.fit(X_train,y_train)
```

```python
r_model.score(X_test,y_test)
```

```python
#plotting model predictions
y_pred = r_model.predict(X_test)
plt.figure(figsize=(12,8))
plt.scatter(y_pred,y_test,c = "green")
```

```python
#optimizing depth of tree
param_gridfr = {"max_depth" : [ 5,15], "n_estimators" :[2,5,10]}
```

```python
#fitting model
gridrfr = GridSearchCV(r_model,param_gridfr)
```

```python
gridrfr.fit(X_train,y_train)
```

```python
#generating predictions
randomforest_predictions = gridrfr.predict(X_test)
```

```python
Model_results(randomforest_predictions)
```

```python
model_names = ["AdaBoost", "KNN", "Random Forest", "Decision Tree", "SVM", "Logistic Regression"]

# Mean Absolute Error (MAE) values
mae_values = [0.4904, 0.5296, 0.5113, 0.3628, 0.4941, 0.4841]

# Root Mean Square Error (RMSE) values
rmse_values = [0.6082, 0.6611, 0.6417, 0.4615, 0.6178, 0.6021]

# Creating subplots for different types of graphs
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

# Bar Chart for MAE
axs[0, 0].bar(model_names, mae_values, color='blue', alpha=0.7)
axs[0, 0].set_title("Bar Chart - MAE")

# Line Graph for MAE
axs[0, 1].plot(model_names, mae_values, marker='o', linestyle='-', color='b')
axs[0, 1].set_title("Line Graph - MAE")

# Scatter Plot for MAE
axs[0, 2].scatter(model_names, mae_values, color='b', marker='o')
axs[0, 2].set_title("Scatter Plot - MAE")

# Bar Chart for RMSE
axs[1, 0].bar(model_names, rmse_values, color='red', alpha=0.7)
axs[1, 0].set_title("Bar Chart - RMSE")

# Line Graph for RMSE
axs[1, 1].plot(model_names, rmse_values, marker='o', linestyle='-', color='r')
axs[1, 1].set_title("Line Graph - RMSE")

# Scatter Plot for RMSE
axs[1, 2].scatter(model_names, rmse_values, color='r', marker='o')
axs[1, 2].set_title("Scatter Plot - RMSE")

# Adjust layout and display
plt.tight_layout()
plt.show()
```
